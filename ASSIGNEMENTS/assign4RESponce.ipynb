### 4.1 Add Two Columns
- Create a new column mpg_hp_sum = mpg + hp using a lambda function.

```python
#mpg_hp_sum = ['mpg','hp'
df['mpg_hp_sum']=df.apply(lambda row: row['mpg'] + row['hp'], axis=1)
df.head()
```

```python
# creat mpg_lable column
df['mpg_label']= df['mpg'].apply(lambda x: 'High' if x>20 else 'Low')
df.head(2)
```
